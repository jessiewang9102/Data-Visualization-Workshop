#### Import packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Jupyter Notebook Magic command to make matplotlib figures show inline

```python
%matplotlib notebook
```

#### Read data from file and view the variable names

```python
data = np.load('cardata.npz', encoding='bytes')
data.files
```

#### Load the variables into arrays

```python
Acceleration = data['Acceleration']
Horsepower = data['Horsepower']
MPG = data['MPG']
Weight = data['Weight']
```

#### Plot data

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.scatter(Horsepower, MPG)
fig1.show()
```

#### Add Accelaration values to shape size

```python
fig2 = plt.figure()
ax2 = fig2.add_subplot(111)
ax2.scatter(Horsepower, MPG, s=Acceleration*12)
fig2.show()
```

#### Add Weight values to marker color

```python
fig3 = plt.figure()
ax3 = fig3.add_subplot(111)
cax = ax3.scatter(Horsepower, MPG, s=Acceleration*12, c=Weight, cmap='copper')
cbar = fig3.colorbar(mappable=cax)
fig3.show()
```

#### Add axis labels and modify marker properties

```python
fig4 = plt.figure()
ax4 = fig4.add_subplot(111)
ax4.set_xlabel('Horsepower')
ax4.set_ylabel('MPG')
ax4.set_title('Car Data')
ax4.grid(linestyle=':', linewidth=0.5)
ax4.set_axisbelow(True)
cax = ax4.scatter(Horsepower, MPG, s=Acceleration*12, c=Weight, cmap='copper', alpha=0.6, edgecolor='none')
cbar = fig4.colorbar(mappable=cax)
cbar.ax.set_title('Weight')
fig4.show()
```
